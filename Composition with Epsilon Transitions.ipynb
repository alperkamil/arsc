# Composition with Epsilon Transitions

```python
import pynini as pn
syms = pn.SymbolTable()
syms.add_symbol('e')  # Epsilon (empty character)
syms.add_symbol('a')
syms.add_symbol('b')
syms.add_symbol('c')
```

### An example composition without epsilon transitions

```python
fst1 = pn.Fst()
fst1.add_state()
fst1.set_start(0)
fst1.set_final(0)
fst1.add_state()
fst1.set_final(1)
fst1.add_arc(0,pn.Arc(1,3,0,1))
fst1.set_input_symbols(syms).set_output_symbols(syms)
```

```python
fst2 = pn.Fst()
fst2.add_state()
fst2.set_start(0)
fst2.set_final(0)
fst2.add_state()
fst2.set_final(1)
fst2.add_arc(0,pn.Arc(3,2,0,1))
fst2.set_input_symbols(syms).set_output_symbols(syms)
```

```python
pn.compose(fst1,fst2)
```

### With epsilon transitions

```python
fst1 = pn.Fst()
fst1.add_state()
fst1.set_start(0)
fst1.set_final(0)
fst1.add_state()
fst1.set_final(1)
fst1.add_arc(0,pn.Arc(1,0,0,1))
fst1.set_input_symbols(syms).set_output_symbols(syms)
```

```python
fst2 = pn.Fst()
fst2.add_state()
fst2.set_start(0)
fst2.set_final(0)
fst2.add_state()
fst2.set_final(1)
fst2.add_arc(0,pn.Arc(0,2,0,1))
fst2.set_input_symbols(syms).set_output_symbols(syms)
```

```python
pn.compose(fst1,fst2)
```
